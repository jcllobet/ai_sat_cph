## Dataset

This data set is related to the direct marketing campaigns of a Portuguese banking institution. Marketing campaigns were based on phone calls. Often, more than one contact with the same client was required, to be able to access whether the product (term bank deposit) would be ('yes') or not ('no') subscribed.

The objective is to predict whether the client will subscribe (yes/no) to a term deposit, building a classification model using decision trees.
